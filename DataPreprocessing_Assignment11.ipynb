**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
```

**Step 2: Importing dataset**

```python
df = pd.read_csv('/content/Data.csv')
df
```

**Step 3: Handling the missing data**

```python
df['Age'].fillna(int(df['Age'].mean()), inplace=True)
df['Salary'].fillna(int(df['Salary'].mean()), inplace=True)
df
```

```python
df['Purchased'].replace(['No', 'Yes'],
                        [0, 1], inplace=True)
df
```

**Step 4: Encoding categorical data**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.Country = le.fit_transform(dfle.Country)
dfle
```

**Step 5: Creating a dummy variable**

```python
dummies = pd.get_dummies(df.Country)
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['Country', 'France'], axis='columns')
final
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
 X = final[["Age", "Salary", "Germany", "Spain"]]
 y = final["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

**Step 7: Feature Scaling**

```python
from sklearn.preprocessing import MinMaxScaler

norm = MinMaxScaler().fit(X_train)

X_train_norm = norm.transform(X_train)

X_test_norm = norm.transform(X_test)
```
